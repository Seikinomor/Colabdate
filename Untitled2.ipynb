```python
from geopy.geocoders import Nominatim

# Ініціалізація геокодувальника
geolocator = Nominatim(user_agent="my-app")

# Список областей України
regions_ukraine = [
    "Вінницька область",
    "Волинська область",
    "Дніпропетровська область",
    "Донецька область",
    "Житомирська область",
    "Закарпатська область",
    "Запорізька область",
    "Івано-Франківська область",
    "Київська область",
    "Кіровоградська область",
    "Луганська область",
    "Львівська область",
    "Миколаївська область",
    "Одеська область",
    "Полтавська область",
    "Рівненська область",
    "Сумська область",
    "Тернопільська область",
    "Харківська область",
    "Херсонська область",
    "Хмельницька область",
    "Черкаська область",
    "Чернівецька область",
    "Чернігівська область"
]

# Отримання координат для кожної області
for region in regions_ukraine:
    try:
        # Отримання геокоду для області
        geocode = geolocator.geocode(region + ", Україна")
        latitude = geocode.latitude
        longitude = geocode.longitude
        print(f"Область: {region}, Широта: {latitude}, Довгота: {longitude}")
    except:
        print(f"Не вдалося знайти координати для області: {region}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from dateutil.parser import parse

# Підключення до Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Завантаження датасету
data_path = '/content/drive/MyDrive/cleaned_data.csv'
df = pd.read_csv(data_path)

# Словник для відповідності українських назв місяців та числових значень
months_dict = {
    'січня': '01',
    'лютого': '02',
    'березня': '03',
    'квітня': '04',
    'травня': '05',
    'червня': '06',
    'липня': '07',
    'серпня': '08',
    'вересня': '09',
    'жовтня': '10',
    'листопада': '11',
    'грудня': '12'
}

# Перетворення стовпця 'created_at' у тип datetime
df['created_at'] = df['created_at'].apply(lambda x: ' '.join(months_dict.get(word, word) for word in x.split()))
df['created_at'] = df['created_at'].apply(lambda x: parse(x, fuzzy=True))

# Виведення перших кількох рядків датасету
#print(df.head())

# Описові статистики
print("Описові статистики:")
print(df.describe())

# Кількість рядків для кожного регіону
region_counts = df['region'].value_counts()

# Графік кількості рядків за регіонами
plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar')
plt.xlabel('Регіон')
plt.ylabel('Кількість')
plt.title('Кількість рядків за регіонами')
plt.show()

# Вибір тільки числових колонок для кореляційної матриці
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
print("Кореляційна матриця:")
print(correlation_matrix)

# Heatmap кореляційної матриці
fig = px.imshow(correlation_matrix, labels=dict(x="Ознаки", y="Ознаки", color="Кореляція"), x=correlation_matrix.columns, y=correlation_matrix.columns)
fig.update_layout(title="Матриця кореляції")
fig.show()

# Графік розподілу даних у колонці 'price'
fig = px.histogram(df, x='price', nbins=30, marginal="rug", title="Розподіл цін")
fig.show()

# Діаграма розсіювання ціни та площі квартири
fig = px.scatter(df, x='price', y='apartment_area_m3', title='Ціна vs Площа')
fig.show()

# Ящик з вусами ціни за різними районами
fig = px.box(df, x='region', y='price', title='Розподіл цін за районами')
fig.show()

# Групування та порівняння середньої ціни за типом будівлі
mean_price_by_building_type = df.groupby('building_type')['price'].mean().reset_index()
fig = px.bar(mean_price_by_building_type, x='building_type', y='price', title='Середня ціна за типом будівлі')
fig.update_layout(yaxis=dict(title="Середня ціна"), xaxis=dict(title="Тип будівлі"))
fig.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from IPython.display import display, HTML, Javascript
import json

# Завантаження датасету
data_path = '/content/drive/MyDrive/cleaned_data.csv'
df = pd.read_csv(data_path)

# Видалення пробілів зі значень у стовпці 'apartment_area_m3'
df['apartment_area_m3'] = df['apartment_area_m3'].str.replace(' ', '')

# Конвертація значень у стовпці 'apartment_area_m3' в числовий тип даних
df['apartment_area_m3'] = pd.to_numeric(df['apartment_area_m3'], errors='coerce')

# Очистка значень у стовпці 'rooms_number'
df['rooms_number'] = df['rooms_number'].str.replace('+', '')

# Конвертація значень у стовпці 'rooms_number' в числовий тип даних
df['rooms_number'] = pd.to_numeric(df['rooms_number'], errors='coerce')

# Видалення рядків з відсутніми значеннями
df.dropna(inplace=True)

# Вибір змінних для моделі
selected_features = ['apartment_area_m3', 'rooms_number', 'nth_floor']

# Розділення датасету на навчальну і тестову вибірки
X = df[selected_features]
y = df['price']

# Побудова та навчання моделі
model = LinearRegression()
model.fit(X, y)

# HTML та JavaScript для інтерфейсу користувача
html = '''
<!DOCTYPE html>
<html>
<head>
    <title>Прогнозування ціни нерухомості</title>
</head>
<body>
    <h1>Прогнозування ціни на оренду та продаж квартир</h1>
    <form id="predictionForm">
        <label for="operation">Тип операції:</label>
        <select id="operation" name="operation">
            <option value="Оренда">Оренда</option>
            <option value="Продаж">Продаж</option>
        </select><br><br>
        <label for="area">Площа квартири (м²):</label>
        <input type="number" id="area" name="area" min="10" max="500" value="80"><br><br>
        <label for="rooms">Кількість кімнат:</label>
        <input type="number" id="rooms" name="rooms" min="1" max="10" value="3"><br><br>
        <label for="floor">Поверх:</label>
        <input type="number" id="floor" name="floor" min="1" max="50" value="1"><br><br>
        <input type="button" value="Прогнозувати" onclick="makePrediction()">
    </form>
    <h2 id="result"></h2>
    <script>
        async function makePrediction() {
            const operation = document.getElementById('operation').value;
            const area = document.getElementById('area').value;
            const rooms = document.getElementById('rooms').value;
            const floor = document.getElementById('floor').value;

            const response = await google.colab.kernel.invokeFunction(
                'notebook.predict',
                [operation, area, rooms, floor],
                {}
            );

            const predictedPrice = response.data['text/plain'];
            document.getElementById('result').innerText = `Прогнозована ціна на квартиру (${operation}): ${predictedPrice}`;
        }
    </script>
</body>
</html>
'''

# Функція для прогнозування ціни
def predict(operation, area, rooms, floor):
    new_data = pd.DataFrame({
        'apartment_area_m3': [float(area)],
        'rooms_number': [int(rooms)],
        'nth_floor': [int(floor)]
    })
    predicted_price = model.predict(new_data)
    return predicted_price[0]

# Реєстрація функції як колаб-функції
from google.colab import output
output.register_callback('notebook.predict', predict)

# Відображення HTML
display(HTML(html))
```
